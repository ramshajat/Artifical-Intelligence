```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
data_file_path = "C:\\Users\\Rimsha\\Desktop\\CS6\\Artifical intelligence Lab\\BCancer.csv"
data_df = pd.read_csv(data_file_path)
data_df.head()
y = data_df["diagnosis"].values
x = data_df.drop(["diagnosis","id","Unnamed: 32"],axis=1)
#print(y)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)# fit label and return encoded labels
print(y)
```

```python
#Divide into training and test data
train_score = []
test_score = []
k_vals = []
tst_size=np.arange(start=0.1, stop=0.3, step=0.1,dtype=float)
for i in range(len(tst_size)):
    train_score = []
    test_score = []
    k_vals = []
    for k in range(1, 21,2):
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = tst_size[i])
        k_vals.append(k)
        knn = KNeighborsClassifier(n_neighbors = k)
        knn.fit(X_train, y_train)
        tr_score = knn.score(X_train, y_train)
        train_score.append(tr_score)
        te_score = knn.score(X_test, y_test)
        test_score.append(te_score)

    plt.figure(figsize=(10,5))
    plt.grid()
    plt.xlabel('Different Values of K')
    plt.ylabel('Model score')
    plt.plot(k_vals, train_score, color = 'r', label = "training score")
    plt.plot(k_vals, test_score, color = 'b', label = 'test score')
    plt.legend(bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure)
    plt.show()
    
    varience = list()
    for item1, item2 in zip(train_score,test_score):
        item = item1 - item2
        varience.append(item)
    plt.figure(figsize=(10,5))
    plt.grid()
    plt.xlabel('Different Values of K')
    plt.ylabel('Actual Varience')
    plt.plot(k_vals,varience, color = 'b', label = 'test score')
    plt.legend(bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure)
    plt.show()

    min_var=min(varience)
    index=varience.index(min_var)
    print("The minimum varience is at {} k value".format(k_vals[index]))
```
